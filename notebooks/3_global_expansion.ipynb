## Chapter 3: Platform Wars – Netflix vs. the Competition

As Netflix grew into a global streaming leader, it wasn’t alone. Competitors like Hulu, Amazon Prime Video, and Disney+ emerged and expanded, creating a fierce battle for viewer attention and market dominance. In this chapter, we compare Netflix’s content library with those of its major rivals to understand how it has maintained an edge in the streaming wars.

```python
# Imports
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Settings
sns.set(style='whitegrid')
plt.rcParams['figure.figsize'] = (10, 6)

# Load dataset
df = pd.read_csv('../data/tv_shows.csv')
df.head()

```

```python
df['Netflix'].value_counts()


```

```python
platform_counts = df[['Netflix', 'Hulu', 'Prime Video', 'Disney+']].sum().sort_values(ascending=False)

platform_counts.plot(kind='bar', color=sns.color_palette("pastel"))
plt.title("Total Number of Titles per Platform")
plt.ylabel("Number of Titles")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

### Total Content Across Platforms

The chart below visualizes the total number of titles available on each streaming platform. Netflix clearly leads the pack, offering the largest selection of titles, followed by Prime Video, Hulu, and Disney+. This breadth of content has played a crucial role in attracting a broad audience base.

```python
yearly_platforms = df.groupby('Year')[['Netflix', 'Hulu', 'Prime Video', 'Disney+']].sum()

yearly_platforms.plot()
plt.title("Platform Title Growth Over Time")
plt.ylabel("Number of Titles")
plt.xlabel("Year")
plt.grid(True)
plt.tight_layout()
plt.show()
```

### Growth of Content Over Time

Next, we look at how each platform has expanded its content library year over year. This temporal comparison helps reveal which platforms are scaling rapidly and which have taken a more conservative approach. Netflix has shown consistent growth, reflecting its aggressive investment in original and licensed content.



```python
def plot_platform_distribution(df):
    """Plots a pie chart of the number of titles available on each streaming platform."""
    platform_sums = df[['Netflix', 'Hulu', 'Prime Video', 'Disney+']].sum()
    platform_sums.index = ['Netflix', 'Hulu', 'Prime Video', 'Disney+']

    colors = sns.color_palette('pastel')[:4]
    plt.pie(platform_sums, labels=platform_sums.index, autopct='%1.1f%%', startangle=140, colors=colors)
    plt.title('Distribution of TV Shows by Streaming Platform')
    plt.tight_layout()
    plt.show()

# Usage
plot_platform_distribution(df)
```

### Distribution of Movies & TV Shows Across Platforms

Finally, the pie chart above illustrates how TV shows are distributed across the four major streaming services. While Netflix still commands the largest share, the competition is steadily gaining ground. This fragmented landscape underscores how viewers today have more options than ever before.

### Summary

This chapter reveals how Netflix continues to lead the streaming landscape with the largest and fastest-growing library of content. However, the rise of Hulu, Prime Video, and Disney+ shows that the competition is intensifying. While Netflix's early investments in original programming and international expansion gave it a head start, other platforms are quickly catching up by building diverse catalogs and catering to niche audiences.
